```
from pyspark.sql import *
```

##### a) Create HR_Employee Pysaprk dataFrame.

```
HR_Employee = spark.read.csv("dbfs:/FileStore/tables/HR_Employee.csv", header=True, inferSchema=True)
```

##### b) Show printSchema & show 10 rows of Dataset

```
HR_Employee.printSchema()
HR_Employee.show(10, truncate=False)
```

##### c) Most employee is working in which department

```
HR_Employee.createTempView("EmployeeTable")
```

```
spark.sql("select Department, count(*) as maxcount from EmployeeTable group by Department order by maxcount desc").show(1)
```

##### d) Highest number of job roles

```
spark.sql("select JobRole, count(*) as maxcount from EmployeeTable group by JobRole order by maxcount desc").show(1)
```

##### e) Which gender have higher strength as workforce?

```
spark.sql("select Gender, count(*) as strength from EmployeeTable group by Gender order by strength desc").show(1)
```

##### f) Compare the marital status of employee and find the most frequent status.

```
spark.sql("select MaritalStatus, count(*) as totalCount from EmployeeTable group by MaritalStatus order by totalCount desc").show(1)
```

##### g) Mostly hired employee have qualification

```
spark.sql("select Education, count(*) as totalcount from EmployeeTable group by Education order by totalcount desc").show(1);
```

##### h) Find the count of employee from which education fields

```
spark.sql("select EducationField, count(*) as totalcount from EmployeeTable group by EducationField order by totalcount desc ").show()
```

##### i) What is the job satisfaction level of employee?

```
spark.sql("select JobSatisfaction, count(*) as totalCount from EmployeeTable group by JobSatisfaction order by totalCount desc ").show(1)
```

##### j) Does most of employee do overtime: Yes or No?

```
spark.sql("select Overtime, count(*) totalCount from EmployeeTable group by Overtime order by totalCount desc").show(1)
```

##### k) Find Min & Max Salaried employees.

```
spark.sql("select max(Income) as Maximum_Salary, min(Income) as Minimum_Salary from EmployeeTable").show()
```

##### l) Does most of the employee do business travel? Find of the employees counts for each category

```
spark.sql("select BusinessTravel, count(*) as totalCount from EmployeeTable group by BusinessTravel order by totalCount desc").show()
```

##### m) Find the AVG salary of graduate employee.

```
spark.sql("select Education, AVG(Income) from EmployeeTable where Education='Bachelor' group by Education").show()
```

##### n) Find the employee qualification receiving salary lower than equal to avg. salary of all employee.

```
spark.sql("select Education, Sum(Income), count(*) as totalcount from EmployeeTable where Income<=(select AVG(Income) from EmployeeTable) group by Education order by totalcount desc").show()
```

##### o) When does the employee have highest chance of promotion in terms of working year?

```
spark.sql("select Performance_Rating, count(*) as PromotionChance from employeetable group by Performance_Rating order by PromotionChance desc ").show()
```

##### p) Highest attrition is in which department?

```
spark.sql("select Department, count(*) as totalcount from EmployeeTable where Attrition='Yes' group by Department order by totalcount desc ").show(1)
```

##### q) Show marital status of Person having highest attrition rate

```
spark.sql("select MaritalStatus, count(*) as totalcount from EmployeeTable where Attrition='Yes' group by MaritalStatus order by totalcount desc ").show(1)
```
